```python
import pandas as pd

# Define file paths
file_paths = [
    ('YOLOv5l', '/content/drive/MyDrive/Lung_Tumor_Detection/runs/train/YOLOv5_VinDr-CXR/results.csv'),
    ('YOLOv7-af', '/content/drive/MyDrive/Lung_Tumor_Detection/runs/train/YOLOv7_VinDr-CXR/results.csv'),
    ('YOLOv8l', '/content/drive/MyDrive/Lung_Tumor_Detection/runs/train/YOLOv8_VinDr-CXR/results.csv'),
    ('YOLOv9c', '/content/drive/MyDrive/Lung_Tumor_Detection/runs/train/YOLOv9_VinDr-CXR/VinDrCXR_run1/results.csv'),
    ('YOLOv10s', '/content/drive/MyDrive/Lung_Tumor_Detection/runs/train/YOLOv10_VinDr-CXR/VinDrCXR_run1/results.csv'),
    ('YOLOv11m', '/content/drive/MyDrive/Lung_Tumor_Detection/runs/train/YOLOv11_VinDr-CXR/VinDrCXR_run1/results.csv'),
    ('YOLOv12m', '/content/drive/MyDrive/Lung_Tumor_Detection/runs/train/YOLOv12_VinDr-CXR/VinDrCXR_run1/results.csv'),
    ('RCS-YOLO', '/content/drive/MyDrive/Lung_Tumor_Detection/runs/train/RCS-YOLO_VinDr-CXR/exp/results.csv'),
    ('CAF-YOLO', '/content/drive/MyDrive/Lung_Tumor_Detection/runs/train/CAF-YOLO_VinDr-CXR/results.csv'),
    ('RepVGG-GELAN', '/content/drive/MyDrive/Lung_Tumor_Detection/runs/train/RepVGG-GELAN_VinDr-CXR/results.csv'),
    ('Proposed Model for X-Rays', '/content/drive/MyDrive/Lung_Tumor_Detection/runs/train/YOLOv5_2_SE_Res_VinDr-CXR/results.csv')
]

# Initialize a list to hold DataFrames
dataframes = []

# Loop through each file path, read the CSV, and extract required columns
for name, path in file_paths:
    df = pd.read_csv(path)
    # Check if all required columns are present
    required_columns = ['   metrics/precision', '      metrics/recall', '     metrics/mAP_0.5', 'metrics/mAP_0.5:0.95']
    if all(col in df.columns for col in required_columns):
        # Rename columns to include experiment name
        renamed_df = df[required_columns].rename(columns=lambda col: f"{name}_{col}")
        dataframes.append(renamed_df)
    else:
        print(f"Missing required columns in {name}")

# Combine all dataframes into a single DataFrame
combined_df = pd.concat(dataframes, axis=1)

# Display the resulting DataFrame
print(combined_df.head())

# Optionally: Save the combined DataFrame if needed
# combined_df.to_csv('/content/combined_metrics.csv', index=False)
```

```python
import matplotlib.pyplot as plt

# Define custom colors for each experiment
custom_colors = {
    'YOLOv5l': '#fbd481',
    'YOLOv7-af': '#c4c465',
    'YOLOv8l': '#88b255',
    'YOLOv9c': '#419f52',
    'YOLOv10s': '#008a58',
    'YOLOv11m': '#007463',
    'YOLOv12m': '#007c85',
    'RCS-YOLO': '#0082bf',
    'CAF-YOLO': '#687dca',
    'RepVGG-GELAN': '#a575c3',
    'Proposed Model for X-Rays': 'red'
}

# List of metrics to plot
metrics = ['   metrics/precision', '      metrics/recall', '     metrics/mAP_0.5', 'metrics/mAP_0.5:0.95']

# Create a figure and axes for a 2x2 grid layout
fig, axes = plt.subplots(2, 2, figsize=(12, 8))  # Smaller individual plots

# Flatten the axes array for easier indexing
axes = axes.flatten()

# Loop through each metric and plot in the corresponding subplot
for i, metric in enumerate(metrics):
    ax = axes[i]  # Get the corresponding subplot

    # Loop through each column in combined_df corresponding to this metric
    for col in combined_df.columns:
        if col.endswith(metric):  # Match exact metric name at the end of the column name
            experiment_name = col.split('_')[0]  # Extract the experiment name
            color = custom_colors.get(experiment_name, 'black')  # Get the color for this experiment
            linewidth = 1.5 if experiment_name == 'RepVGG-GELAN + 3 Fire + CBAM (proposed)' else 1
            ax.plot(combined_df.index, combined_df[col], label=experiment_name, color=color, linewidth=linewidth)

    # Add labels, title, and grid to the subplot
    ax.set_xlabel('Epoch')
    ax.set_ylabel(metric.split('/')[-1].replace('_', ' ').capitalize())
    ax.set_title(f'{metric.split("/")[-1].replace("_", " ").capitalize()}')
    ax.grid(True)

    # Add legend to the 3rd subplot (index 2) only
    if i == 2:
        ax.legend()

# Adjust layout for better appearance
plt.tight_layout()

# Show the plots
plt.show()
```
